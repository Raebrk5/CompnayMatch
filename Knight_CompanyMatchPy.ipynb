```python
'''This code was created to enable anyone with a python IDE, reguardless of access to BI's proprietary tools, 
be able to match the DCA SQL chron results to a vertical without knowledge of the DCA brands. 
To use this you will need to have access to the daily chron results & the live document of brands for the DCA
team. For access to these components please conatact LeGO associate, B. Knight (bknight@indeed.com)

Use: 
1)download the usDcaDailyRun_[DATE].tsv from the chron results to the same folder as this file 
(please ensure they are in the same file - this is so that the data can be removed after matched
this is to follow the Indeed PII regulations)
2) ensure it is the same day information [amend the file name on line 21 & comment out 20]
3) Run! 
Your newly created doc ('US-S.csv' in the root folder) will contain the information for upload!
'''

import pandas as pd
import re
import datetime as dt

#format the daily file name (done daily so the date is = to the current date) for the SQL results
daily_file_date = dt.datetime.now().strftime("%Y_%m_%d")
daily_file = "usDcaDailyRun_"+daily_file_date+".tsv"

#reads the index list of brands to verticals (an external dictionary)
#pulls in the brand field and creates a compiled list of brands 
vertical_list = pd.read_table("brands_verticals.tsv")
vert_list_str = str(vertical_list['brand'])
normtitle_list = tuple(vertical_list['brand'])
test_list = re.compile('|'.join(normtitle_list))


#pulls in the daily SQL result file to match a lower case version of the company names 
#creates a new row for them
pDCA = pd.read_table(daily_file)
pDCA.head()
pDCA['daily_list'] = pDCA['company'].str.lower()

#creates a function for iterating through the daily_list seraching for partial matches to the company names
def re_search(x):
    match = re.search(test_list, x)
    if match:
        return match.group(0)

#add column to daily df and adds a column for the vertical match (type) to the brand with function defined above
pDCA['brand match'] = pDCA['daily_list'].apply(re_search)

# merges results to new csv 'us-s'
company_list_merged = pd.merge(pDCA, vertical_list, how='left', left_on='brand match', right_on='brand')[['account_id','daily_list','brand','vertical']]
company_list_merged['vertical'].fillna('#N/A', inplace=True)
company_list_merged.to_csv('US-S.csv', encoding='utf-8')

#print the results to check for error 
company_list_merged
```
